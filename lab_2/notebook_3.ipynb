### Создание экземпляра SparkSession

```python
spark = SparkSession.builder.appName('Homework_2_lab3').getOrCreate()
```

### Задание 1. Считать файл 1.csv и создать Spark DataFrame

```python
df = spark.read.format('csv').load('data/3.csv')
df.show(5)
```

```python
def prepare_data(row):
  result = []
  for el in row:
    if el != 'null':
      result.append(int(el))
    else:
      result.append(None)

  return result

df = df.rdd.map(prepare_data).collect()
df = spark.createDataFrame(df)

df.show(5)
```

### Задание 2. Подсчитать количество пропусков

```python
columns_na_count = df.select([count(when(isnull(c), c)).alias(c) for c in df.columns]).collect()
columns_na_count
```

### Задание 3. Обработать пропуски: заполнить пропуски средним по колонке

```python
column_avg_values = df.select([avg(c).alias(c) for c in df.columns]).collect()
column_avg_values
```

```python
to_fill = {}

for column in range(len(columns_na_count[0])):
  if columns_na_count[0][column] > 0:
    to_fill[f'_{column+1}'] = column_avg_values[0][column]

to_fill
```

```python
df_filled = df.na.fill(to_fill)
df_filled.show(5)
```

### Задание 4. Произвести фильтрацию данных: оставить строки, сумма которых < 0

```python
df_filtered = df_filled \
  .withColumn('avg', expr(f'({"+".join([c for c in df_filled.columns])}) / {str(len(df_filled.columns))}')) \
  .where("avg < 0")

df_filtered.show(5)
```

### Задание 5. Подсчитать статистику по каждой колонке и сохранить в отдельный набор данных

```python
df_stats = df_filtered.describe()
df_stats.show()
```

### Задание 6. Записать полученный набор данных со статистикой в r_3.csv

```python
df_stats.write.json('data/r_3.json', mode='overwrite')
```
